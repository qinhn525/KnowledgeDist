```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("./engineer.tsv", sep="\t")
print(len(data))
data.head()
```

```python
data_group = data.groupby("discipline")
print(len(data_group))
```

```python
lens = [len(item[1]) for item in data_group]
print(lens)
plt.plot([x for x in range(len(lens))], lens)
```

```python
# 每个类挑5个出来作为test数据
test = [item[1].sample(50) for item in data_group]
test_out = pd.concat(test)
print(len(test_out))
test_out.to_csv("./engineer_test.tsv", sep="\t", index=False, columns=["abstract", "discipline"])
```
